```python
#hangman library
hangman={0:'''
        ____________
         |



         ''',
        1:'''
        ____________
         |
         O


         ''',
        2:'''
        ____________
         |
         O
        /

        ''',
        3:'''
        ____________
         |
         O
        / \\

        ''',
        4:'''
        ____________
         |
         O
        / \\
         |
         ''',
        5:'''
        ____________
         |
         O
        / \\
         |
        /''',
        6:'''
        ____________
         |
         O
        / \\
         |
        / \\ '''}
```

```python
import random
list_of_words = ['python', 'colab', 'data']

# the secret word in capital letters is randomly picked out of the list_of_words
answer_word = random.choice(list_of_words).upper()

# number of underscores which is the length of secret word
hidden_word = "_ " * len(answer_word)

# number of how many mistakes player can do
remaining_mistakes = 6
print('Welcome to the hangman game!')
print(hangman[0])



# loops the game. condition for looping is that variable that mistakes are remaining and there are still underscores left after guessed letter
while remaining_mistakes > 0 and hidden_word.count("_") != 0:

  # statement, which shows the underscores/ lenght of secret word
  print(f'\nThe hidden word is {hidden_word}. Pick a letter:\n')

  # user is asked to pick a letter while ensuring input is uppercase
  picked_letter = input().upper()

  # counts the indec of which underscore in hidden_word corresponds with picked_letter
  index_count = 0


  # loop for each letter in answer_word
  for letter in answer_word:

    # checks picked_letter against each letter in answer_word
    if picked_letter == letter:

      # slices and puts the hidden_wordn together and replaces the underscore with the right guessed picked_letter on the right index
      hidden_word = hidden_word[0:index_count] + letter + hidden_word[index_count+1:]

    # goes to next underscore. 2 steps because of space after underscore
    index_count += 2


  # looking for if the picked letter is in the secret word
  if picked_letter in answer_word:

    # printing this, if the hidden word contains picked_letter
    print(f'\nThats correct, the word contains the letter {picked_letter}!')

  else:

    # if wrong attempt, one lesser remaining mistake
    remaining_mistakes -= 1

    print(f'\nWRONG! \nNumber of mistakes left: {remaining_mistakes}\n')
    print(hangman[6-remaining_mistakes])

  print('\n======================================================')

# when no remaining guess is left, you've lost
if remaining_mistakes == 0:

  print('HANGED!!!!!!!')

# printing this, when you_ve won
else:
  print('CONGRATULATIONS!')

# prints the hidden word at the end
print(f'The hidden word is "{answer_word}"!')


```
